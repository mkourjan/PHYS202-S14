Question 1

For this question refer to the solutions for the simple pendulum we got using scipy's odeint from the beginning of the ODE Tour.

(a) See what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs t for 5 different initial angles, θ0 , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 and π ?

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.integrate import odeint

g = 9.8 
ell = 0.5

th0 = 0.*pi/180 
th45 = 45.*pi/180
th90 = 90.*pi/180
th135 = 135.*pi/180
th180 = 180*pi/180

t = np.arange(0.,10.+0.01,0.01)

theta_app0 = lambda t: th0*np.cos(np.sqrt(g/ell)*t)
theta_app45 = lambda t: th45*np.cos(np.sqrt(g/ell)*t)
theta_app90 = lambda t: th90*np.cos(np.sqrt(g/ell)*t)
theta_app135 = lambda t: th135*np.cos(np.sqrt(g/ell)*t)
theta_app180 = lambda t: th180*np.cos(np.sqrt(g/ell)*t)

def deriv(theta_current,t):  
    g = 9.8 
    ell = 0.5 
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])

t = np.arange(0.,10.+0.01,0.01)

omega0 = 0. 

theta_init0 = np.array([th0, omega0]) 
theta_init45 = np.array([th45, omega0])
theta_init90 = np.array([th90, omega0])
theta_init135 = np.array([th135, omega0])
theta_init180 = np.array([th180, omega0])

theta0 = odeint(deriv,theta_init0,t)
theta45 = odeint(deriv,theta_init45,t) 
theta90 = odeint(deriv,theta_init90,t) 
theta135 = odeint(deriv,theta_init135,t) 
theta180 = odeint(deriv,theta_init180,t) 

plt.figure(figsize = (20, 20))

plt.plot(t,theta_app0(t),'r-',label="Approximate solution - 0")
plt.plot(t,theta0[:,0],'r--',label="Numerical solution - 0")

plt.plot(t,theta_app45(t),'y-',label="Approximate solution - 45")
plt.plot(t,theta45[:,0],'y--',label="Numerical solution - 45")

plt.plot(t,theta_app90(t),'g-',label="Approximate solution - 90")
plt.plot(t,theta90[:,0],'g--',label="Numerical solution - 90")

plt.plot(t,theta_app135(t),'b-',label="Approximate solution - 135")
plt.plot(t,theta135[:,0],'b--',label="Numerical solution - 135")

plt.plot(t,theta_app180(t),'m-',label="Approximate solution - 180")
plt.plot(t,theta180[:,0],'m--',label="Numerical solution - 180")

plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)

plt.xlim(0, 10)
plt.ylim(-4,5)
plt.grid()
plt.legend()
plt.show()
```

As I increase my angle, the numerical solution, that is represented by the dotted line, also increases!

At theta = 0, the trajectory is just zero because the pendulum is not pulled to any angle and therefore its at rest.  When theta = pi, there is a more interesting scenario.  Our program does not account for the bending of a string, so our program models the trajectory as if the pendulum is stuck in a position where its mass is pointing straight up!

(b) Make plots of θ vs. t, ω vs. t, and ω vs. θ (the so-called "phase plot") for θ0 = 0.95π. Look carefully at their behavior. What is going on in these figures?

```python
g = 9.8 
ell = 0.5

th0 = 0.95*pi 

t = np.arange(0.,10.+0.01,0.01)

omega0 = 0

theta_init0 = np.array([th0, omega0])

theta = odeint(deriv,theta_init0,t)

plt.figure(figsize = (10, 5))


plt.plot(t, theta[:, 1], 'b', label = "omega vs. t")

plt.xlim(0, 10)
plt.ylim(-10,10)
plt.grid()
plt.xlabel('T')
plt.ylabel('Omega')
plt.legend(loc = 'best')
plt.show()

print "In the graph, the angular velocity of the pendulum varies with time.  The pendulum is swinging back and forth and it accelerates on the way down and decelerates on the way up as our graph indicates."
```

```python
plt.figure(figsize = (10, 5))

plt.plot(t,theta[:,0],'r',label="theta vs. t")

plt.xlim(0, 10)
plt.ylim(-10,10)
plt.grid()
plt.xlabel('X')
plt.ylabel('Theta')
plt.legend(loc = 'best')
plt.show()

print "The angle theta is changing with time, but since we defined our starting theta, we only have one trajectory and the pendulum hits the same angle from either side at the peak of each swing."
```

```python
plt.plot(theta[:, 1], theta[:, 0], 'm', label = 'omega vs. theta')
plt.xlim(-10, 10)
plt.ylim(-10,10)
plt.grid()
plt.xlabel('Theta')
plt.ylabel('Omega')
plt.legend(loc = 'best')
plt.show()

print "The angular velocity changes with time.  When omega = 0, the theta must be at its largest because the pendulum is coming to rest and changing direction.  On the other hand when Theta = 0, omega must be at its largest because the pendulum is going through the bottom of the swing when it must be moving fastest."
```

(c) For the θ0 = 0.95π case, check that energy is conserved by plotting E/m vs. t (where E/m=gℓ(1−cos(θ))+12ℓ2ω2 ).
Is the energy a constant? If not, then the computational errors are producing a non-physical loss (or gain) of energy. Review the documentation for odeint and change the error tolerance value to a tighter number.

```python
theta1 = odeint(deriv, theta_init0, t,)

Em = g * ell * (1 - np.cos(theta1[:, 0])) + (0.5 * ell ** 2 * theta1[:, 1] ** 2)
t = np.arange(0.,10.+0.01,0.01)

plt.figure(figsize = (10, 5))
plt.grid()
plt.plot(t, Em, label = 'E/m vs. t')
plt.ylim(0,15)
plt.xlabel('t')
plt.ylabel('E/m')
plt.legend(loc = 'best' )
plt.show

print "The energy is constant which is consistant with the situation. In order for a pendulum to maintain the same angle theta at each peak, energy must be conserved and our graph confirms this."
```

Question 2

(a) Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.

dydx=x+2y

Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.

In [5]:

```python
def dydx(dx):
    xmax = 2
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    x[0] = 0
    y[0] = 0
    f = x + 2 * y
    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        
        y[i+1] = y[i] + dx * (x[i] + 2 * y[i])
        
    return x, y
```

(b) Plot the results from part (a) for the two different step sizes with the exact solution
y=0.25e2x−0.5x−0.25 on one plot. Include a legend. Which step size gives a more accurate result?

```python
x1, y1 = dydx(0.25)
x2, y2 = dydx(0.02)

x = np.arange(0.,10.+0.01,0.01)
y = 0.25 * e ** (2 * x) - 0.5 * x - 0.25
plt.figure(figsize = (10, 5))
plt.grid()
plt.xlim(0, 3)
plt.ylim(0, 10)
plt.plot(x, y, color = 'r', label = "exact solution")
plt.plot(x1, y1, color = 'b', label = 'diff equation with dx = 0.25')
plt.plot(x2, y2, color = 'g', label = 'diff equqtion with dx = 0.02')
plt.legend(loc = 'best')
plt.xlabel('X')
plt.ylabel('Y')
plt.show

print "When the step sizes are smaller or 0.02, the result is more accurate."
```

(c) Now solve the same problem using Euler-Cromer and the Midpoint method for the same pair of step sizes and plot the results along with the exact solution. Then also plot the difference between the approximate numerical solutions and the exact solution. Which combination of method and step-size gives the best result?

```python
x1, y1 = dydx(0.25)
x2, y2 = dydx(0.02)

def dydxec(dx):
    xmax = 2
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    x[0] = 0
    y[0] = 0
    f = x + 2 * y
    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        
        y[i+1] = y[i] + dx * (x[i + 1] + 2 * y[i + 1])
        
    return x, y

def dydxm(dx):
    xmax = 2
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    x[0] = 0.
    y[0] = 0.
    f = x + 2 * y
    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        
        y[i+1] = y[i] + dx * (0.5 * ((x[1] + x[i + 1]) + 2 * (y[1] + y[i + 1])))
        
    return x, y
x1, y1 = dydxec(0.25)
x2, y2 = dydxec(0.02)
x3, y3 = dydxm(0.25)
x4, y4 = dydxm(0.02)

x = np.arange(0.,10.+0.01,0.01)
y = 0.25 * np.exp(2 * x2) - 0.5 * x2 - 0.25
plt.figure(figsize = (10, 5))
plt.grid()
plt.xlim(0, 3)
plt.ylim(0, 10)
plt.plot(x2, y, color = 'r', label = "exact solution")
plt.plot(x1, y1, color = 'b', label = 'euler-cromer method - dx = 0.25')
plt.plot(x2, y2, color = 'm', label = 'euler-cromer method - dx = 0.02')
plt.plot(x3, y3, color = 'y', label = 'midpoint method - dx = 0.25')
plt.plot(x4, y4, color = 'g', label = 'midpoint method-dx = 0.02')
plt.legend(loc = 'best')
plt.xlabel('X')
plt.ylabel('Y')
plt.show

```

```python
error = y - y4
error1 = y - y2
plt.figure(figsize = (10, 5))
plt.grid()
plt.xlim(0, 3)
plt.ylim(0, 10)
plt.plot(x4, error, color = 'r', label = "error - midpoint")
plt.plot(x2, error1, color = 'g', label = "error - EulerCromer")

plt.plot(x2, y, color = 'b', label = "original function")
plt.legend(loc = 'best')
plt.xlabel('X')
plt.ylabel ('Error')
plt.show

print "The midpoint method has the best solution since it has a function more similar to the original function than the eulercromer function does."
```

(a) Create a derivative function to solve this system of equations and use odeint to solve it for a projectile fired at an angle of 50 degrees from the horizontal at an initial speed of 30 m/s. Track the projectile's motion over the time interval of 0 to 200 seconds and let (x0,y0) = (0,0). You'll have to calculate (vx0,vy0) from the information provided. Let vter=44 m/s. Plot the trajectory of the projectile (y vs. x) to see how far it can travel in the x direction before landing.

```python
def trajec(i, t, vt):
    g = 9.81
    vx = i[0]
    vy = i[1]
    dvx = -g * (vx/(vt ** 2)) * sqrt((vx ** 2) + (vy ** 2))
    dvy = -g * (1 + (vy/vt**2 * sqrt(vx**2 * sqrt(vx**2 + vy**2))))
    return np.array([dvx,dvy, vx, vy])
def trajectory(v0, theta0, x0, y0, vt):
    t = np.arange(0., 200.01, 0.01)
    vx0 = v0 * np.cos(theta0 * np.pi / 180)
    vy0 = v0 * np.sin(theta0 * np.pi/180)
    i = np.array([vx0, vy0, x0, y0])
    vel = odeint(trajec, i, t, args = (vt, ))
    plt.grid()
    plt.plot(vel[:,2], vel[:,3], label = 'trajectory of projectile')
    plt.xlim(0, 80)
    plt.ylim(0, 20)
    plt.xlabel('x', fontsize = 20)
    plt.ylabel('y', fontsize = 20)
    plt.legend(loc = 'best')
    plt.show()
    
trajectory(30., 50, 0., 0., 44)
```
