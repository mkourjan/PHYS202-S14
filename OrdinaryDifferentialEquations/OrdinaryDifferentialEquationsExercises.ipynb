```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.integrate import odeint

g = 9.8 
ell = 0.5

th0 = 0.*pi/180 
th45 = 45.*pi/180
th90 = 90.*pi/180
th135 = 135.*pi/180
th180 = 180*pi/180

t = np.arange(0.,10.+0.01,0.01)

theta_app0 = lambda t: th0*np.cos(np.sqrt(g/ell)*t)
theta_app45 = lambda t: th45*np.cos(np.sqrt(g/ell)*t)
theta_app90 = lambda t: th90*np.cos(np.sqrt(g/ell)*t)
theta_app135 = lambda t: th135*np.cos(np.sqrt(g/ell)*t)
theta_app180 = lambda t: th180*np.cos(np.sqrt(g/ell)*t)

def deriv(theta_current,t):  
    g = 9.8 
    ell = 0.5 
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])

t = np.arange(0.,10.+0.01,0.01)

omega0 = 0. 

theta_init0 = np.array([th0, omega0]) 
theta_init45 = np.array([th45, omega0])
theta_init90 = np.array([th90, omega0])
theta_init135 = np.array([th135, omega0])
theta_init180 = np.array([th180, omega0])

theta0 = odeint(deriv,theta_init0,t)
theta45 = odeint(deriv,theta_init45,t) 
theta90 = odeint(deriv,theta_init90,t) 
theta135 = odeint(deriv,theta_init135,t) 
theta180 = odeint(deriv,theta_init180,t) 

plt.figure(figsize = (20, 20))

plt.plot(t,theta_app0(t),'r-',label="Approximate solution - 0")
plt.plot(t,theta0[:,0],'r--',label="Numerical solution - 0")

plt.plot(t,theta_app45(t),'y-',label="Approximate solution - 45")
plt.plot(t,theta45[:,0],'y--',label="Numerical solution - 45")

plt.plot(t,theta_app90(t),'g-',label="Approximate solution - 90")
plt.plot(t,theta90[:,0],'g--',label="Numerical solution - 90")

plt.plot(t,theta_app135(t),'b-',label="Approximate solution - 135")
plt.plot(t,theta135[:,0],'b--',label="Numerical solution - 135")

plt.plot(t,theta_app180(t),'m-',label="Approximate solution - 180")
plt.plot(t,theta180[:,0],'m--',label="Numerical solution - 180")

plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)

plt.xlim(0, 10)
plt.ylim(-4,5)
plt.grid()
plt.legend()
plt.show()
```
