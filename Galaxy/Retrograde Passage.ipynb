```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import random
from scipy.integrate import odeint
from scipy.integrate import ode
```

# Retrograde Passage

### Description: Mass of galaxy "S" is equal to the mass of galaxy "M". Also the galaxy "S" approaches in the opposite direction in relation to the direction that the stars in galaxy "M" are orbitting.

```python
gamma = 4.494e-8 #kpc^3/(Solar Mass)/(10^8 years)^2
M = 10**11 #Mass of galaxy "M" in Solar Masses
stars = [12, 18, 24, 30, 36] #How many stars are in each orbit
star_rad = [.2, .3, .4, .5, .6] #Percentage that describes star orbit 
                                #distribution in relation to rmin of approaching galaxy
    
    
rmin = 20 #kpc, refers to minimum distance of galaxy "S" to center of galaxy "M"

'''Create initial coordinates for the stars in galaxy "M" in terms of polar coordinates: 
    r (distance from star to center of "M") and theta (angle of "r" from 0 radians)'''

r0 = np.zeros(120)
theta0 = np.zeros(120)

i = 0
for si, n in enumerate(stars):
    for x in range(n):
        r0[i] = star_rad[si] * rmin
        theta0[i] = np.linspace(0, 2*np.pi, n)[x]
        i+=1
        
'''Create initial velocities for stars in terms of "r" and "theta".'''        

v0 = np.sqrt(gamma*M/r0)
vtheta = theta0+np.pi/2


'''Convert radial velocities and positions into cartesian coordinates.'''

vx0 = v0*np.cos(vtheta)
vy0 = v0*np.sin(vtheta)


x0 = r0 * np.cos(theta0)
y0 = r0 * np.sin(theta0)

'''Plot the star positions and their corresponding velocity vectors.'''

plt.figure(figsize = (6,6))
plt.grid()
plt.scatter(x0, y0)
plt.quiver(x0, y0, vx0, vy0)
plt.show()
```

```python
Rx0 = -120 #initial 'x' position of "S" galaxy
Ry0 = 120 #initial 'y' position of "S" galaxy
dRx0 = 10 #initial 'x' velocity of "S" galaxy
dRy0 = -5 #initial 'y' velocity of "S" galaxy

initarray = np.array([Rx0, Ry0, dRx0, dRy0])

'''Function describing movement of approaching galaxy "S".  Inputs an array of initial position and velocity
values, and outputs an array of new position and velocity values.'''

def R_2ndDeriv(initarray, t):
    M = 10**11 #kpc
    gamma = 4.494e-8 #kpc^3/(Solar Mass)/(10^8 years)^2
    S = M #Mass of approaching galaxy in terms of "M"
    
    Rx = initarray[0]
    Ry = initarray[1]
    dRx = initarray[2]
    dRy = initarray[3]
    
    R = np.sqrt(Rx**2 + Ry**2) #Magnitude of R, where R is distance of "S" from "M"
    
    vRx = -gamma*(M+S)/(R**3)*Rx #differential equation given in prompt
    vRy = -gamma*(M+S)/(R**3)*Ry #differential equation given in prompt
    
    return np.array([dRx, dRy, vRx, vRy])
    
```

```python
t = np.arange(0, 23.1,.1) #time
S = M #Mass of "S" in terms of "M"


gamma = 4.494e-8 #kpc^3/(Solar Mass)/(10^8 years)^2

'''Use the 'odeint' function to solve the differential equation that describes the movement of
the approaching "S" galaxy'''

R = odeint(R_2ndDeriv, initarray, t) 

Rx = R[:, 0]
Ry = R[:, 1]
Rvx = R[:, 2]
Rvy = R[:, 3]

'''Plot the trajectory of the approaching 'S' galaxy'''

plt.plot(Rx, Ry)
plt.grid()
plt.plot(0, 0, 'ro')
plt.show()
```

```python
'''initial conditions for "M" galaxy'''
rx0 = x0[0]
ry0 = y0[0]
drx0 = vx0[0]
dry0 = vy0[0]

'''Reiterate the initial conditions for "S" galaxy'''

Rx0 = -120
Ry0 = 120
dRx0 = 10
dRy0 = -5

'''Put 'M' and 'S' galaxy information into an array'''

initarray = np.array([rx0, ry0, drx0, dry0, Rx0, Ry0, dRx0, dRy0])

'''Function describing the motion of the stars in the "M" galaxy.'''

def r2_dot(t, initarray):
    M = 10**11 #kpc
    gamma = 4.494e-8 #kpc^3/(Solar Mass)/(10^8 years)^2
    S = M #Mass of 'S' galaxy
    
    
    
    rx = initarray[0]
    ry = initarray[1]
    drx = initarray[2]
    dry = initarray[3]
    Rx = initarray[4]
    Ry = initarray[5]
    dRx = initarray[6]
    dRy = initarray[7]
    
    r = np.sqrt(rx**2 + ry**2) #magnitude of r
    R = np.sqrt(Rx**2 + Ry**2) #magnitude of R
    
    px = Rx - rx
    py = Ry - ry
    p = R - r #distance of star to approaching galaxy
    
    '''equations below are given in the problem prompt.  They are differential equations that
    describe the movement of the stars in "M" and the galaxy "S"'''
    
    ddrx = -gamma * ((M/r**3)*rx - (S/p**3)*px+ S/(R**3)*Rx)
    ddry = -gamma * ((M/r**3)*ry - (S/p**3)*py+ S/(R**3)*Ry)
    ddRx = -gamma * (M+S)/(R**3) * Rx
    ddRy = -gamma * (M+S)/(R**3) * Ry
    
    return np.array([drx, dry, ddrx, ddry,dRx, dRy, ddRx, ddRy])
```

```python
'''Use the "ode" function to solve the differential equation for the motion of stars.'''

r = ode(r2_dot).set_integrator('vode', method='bdf',max_step=5000)

start_time = 0.0 #initial time
end_time = 20.0 #final time
delta_time = 1./2. #change in time for each time step

'''number of time steps'''

time_steps = np.floor((end_time - start_time)/delta_time) + 1

'''create empty arrays to add to later.  These are important for their shape.'''

t = np.zeros((time_steps))
x = np.zeros((120,time_steps))
y = np.zeros((120,time_steps))
vx = np.zeros((120,time_steps))
vy = np.zeros((120,time_steps))
Rx1 = np.zeros((time_steps))
Ry1 = np.zeros((time_steps))
vRx1 = np.zeros((time_steps))
vRy1 = np.zeros((time_steps))

'''initilize conditions'''

t[0] = start_time
x[:,0],y[:,0],vx[:,0],vy[:,0] = x0,y0,vx0,vy0
Rx1[0],Ry1[0],vRx1[0],vRy1[0] = Rx0,Ry0,dRx0,dRy0

'''Loop through data and integrate over each timestep'''

for i,(x0i,y0i,vx0i,vy0i) in enumerate(zip(x0,y0,vx0,vy0)):
    initarray = np.array([x0[i],y0i,vx0i,vy0i,Rx0,Ry0,dRx0,dRy0])
    r.set_initial_value(initarray, start_time)
    k = 1
    while r.successful() and k < time_steps:
        r.integrate(r.t + delta_time)
        
        '''The information below is used to plot in the next cell.'''
        t[k] = r.t
        x[i,k] = r.y[0]
        y[i,k] = r.y[1]
        vx[i,k] = r.y[2]
        vy[i,k] = r.y[3]
        Rx1[k] = r.y[4]
        Ry1[k] = r.y[5]
        vRx1[k] = r.y[6]
        vRy1[k] = r.y[7]
        k += 1
```

```python
'''Loop through the information and plot a graph for each timestep.''' 

for i, j in enumerate(t):
    plt.figure(figsize = (5, 5))
    plt.grid()
    plt.title('time = %.1f'%j)
    plt.xlim(-70, 70)
    plt.ylim(-70, 70)
    plt.scatter(x[:, i], y[:, i])
    plt.plot(Rx1[i], Ry1[i], 'ro')
    plt.show()
```

The retrograde passage situation is caused when the approaching galaxy and the rest-frame galaxy are orbiting and moving in opposite directions.  In this situation, the stars and approaching galaxy pass each other at a very fast rate since both objects are speeding past each other.  Since this is the case, the individual stars spend a lot less time in the proximity of the approaching galaxy in comparison with the direct passage situation.  Due to this fact, the approaching galaxy mostly disrupts the outer rings of the rest-frame galaxy, and does not engulf many of the stars.
